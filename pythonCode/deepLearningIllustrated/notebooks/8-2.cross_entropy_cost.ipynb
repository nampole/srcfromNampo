```python
from numpy import log
```

```python
def cross_entropy(y, a):    # 크로스 엔트로피 공식 적용해서 함수를 생성
    return -1*(y*log(a) + (1-y)*log(1-a))
```

```python
cross_entropy(1, 0.9997)
```

```python
cross_entropy(1, 0.9)
```

```python
cross_entropy(1, 0.6)
```

```python
cross_entropy(1, 0.1192)
```

```python
cross_entropy(0, 0.1192)  # 이차비용함수와 비교 해보면 값이 더 크므로 신경망을 더 빠르게 학습시키는 것이다 ???
```

```python
cross_entropy(1, 1-0.1192)
```
