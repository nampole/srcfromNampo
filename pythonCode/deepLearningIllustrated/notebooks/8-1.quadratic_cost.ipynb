```python
def squared_error(y, yhat):     # y:정답 yhat : 예측값
    return (y - yhat)**2
```

```python
squared_error(1, 1)
```

```python
squared_error(1, 0.9997)
```

```python
squared_error(1, 0.9)
```

```python
squared_error(1, 0.6)
```

```python
squared_error(1, 0.1192)
```

```python
squared_error(0, 0.1192)
# 크로스-엔트로피 함수와 비교 해보면 포화영역 값을 입력했을때 결과값이 더 작으므로 신경망을 더 느리게 학습시키는 것이다 ???
```
