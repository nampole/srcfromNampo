```python
'''
Unit converter: Miles and Kilometers
'''
def print_menu():
    print('1. Kilometers to Miles')
    print('2. Miles to Kilometers')
def km_miles():
    km = float(input('Enter distance in kilometers:'))
    miles = km / 1.609
    print('Distance in miles: {0}'.format(miles))
def miles_km():
    miles = float(input('Enter distance in miles: '))
    km =miles * 1.609
    print('Distance in kilometers: {0}'.format(km))
if __name__ == '__main__':
    print_menu()
    choice = input('Which conversion would you like to do?:')
    if choice == '1':
        km_miles()
    if choice == '2':
        miles_km()

print_menu()
km_miles()
```

```python
'''
Quadratic equation root calculator
'''
def roots(a, b, c):
    D = (b*b - 4*a*c)**0.5
    x_1 = (-b + D)/(2*a)
    x_2 = (-b - D)/(2*a) 
    
    print('x1: {0}'.format(x_1))
    print('x2: {0}'.format(x_2))
if __name__ == '__main__':
    a = input('Enter a: ')
    b = input('Enter b: ')
    c = input('Enter c: ')
    roots(float(a), float(b), float(c))

roots (10, 15, 87)
```

### 목록 또는 튜플에 대한 반성

다음과 같이 for 루프를 사용하여 목록이나 튜플을 검토할 수 있습니다:

```python
l = [1, 2, 3]
for item in l:
    print(item)

# 이것은 목록에 있는 항목들을 출력합니다:
```

<h5> Matplotlib을 사용하여 그래프 만들기 </h5>

우리는 Python으로 그래프를 만들기 위해 matplotlib을 사용할 예정입니다. Matplotlib은 Python 패키지이며, 이는 관련 기능을 가진 모듈들의 모음임을 의미합니다. 이 경우, 모듈은 숫자를 플롯하고 그래프를 만드는 데 유용합니다. Matplotlib은 Python의 표준 라이브러리에 내장되어 있지 않으므로 설치하셔야 합니다. 설치 지침은 부록 A에 포함되어 있습니다. 설치가 완료되면 Python 셸을 시작하십시오.
설명된 바와 같이 설치 지침은 IDLE 셸을 계속 사용하거나 Python의 내장 셸을 사용할 수 있습니다.

이제 우리는 첫 번째 그래프를 만들 준비가 되었습니다. 우리는 (1, 2), (2, 4), (3, 6) 세 점만으로 구성된 간단한 그래프부터 시작하겠습니다. 이 그래프를 만들기 위해 먼저 두 개의 숫자 목록을 만들겠습니다—하나는 이러한 점들의 x좌표 값을 저장하고, 다른 하나는 y좌표를 저장합니다. 다음 두 문장은 정확히 그 작업을 수행하여 x_numbers와 y_numbers 두 리스트를 생성합니다:

```python
X_numbers = [1, 2, 3]
y_numbers = [2, 4, 5]
```

여기에서 우리는 플롯을 만들 수 있습니다:

```python
from pylab import plot, show
plot(X_numbers, y_numbers, marker='*')
# plot(X_numbers, y_numbers, 'o')
show()
#[<matplotlib.lines.Line2D object at 0x7f83ac60df10>]
```

첫 번째 줄에서는 matplotlib 패키지의 일부인 pylab 모듈에서 plot() 및 show() 함수를 가져옵니다. 다음으로, 두 번째 줄에서 plot() 함수를 호출합니다. Plot() 함수의 첫 번째 인수는 x축에 플롯하려는 숫자들의 목록이며, 두 번째 인수는 y축에 플롯하고자 하는 해당 숫자들의 목록입니다. Plot() 함수는 객체를 반환합니다 — 보다 정확히 말하면 객체를 포함하는 리스트입니다. 이 객체는 우리가 파이썬에 생성하도록 요청한 그래프에 대한 정보를 포함하고 있습니다. 이 단계에서는 그래프에 제목과 같은 추가 정보를 추가할 수 있거나, 그래프를 그대로 표시할 수 있습니다. 지금은 그래프만 표시하겠습니다.

Plot() 함수는 그래프만 생성합니다. 실제로 표시하려면 show() 함수를 호출해야 합니다:

>>> show()

원점(0, 0)에서 시작하는 대신 x축은 숫자 1에서 시작하고 y축은 숫자 2에서 시작한다는 점에 유의하십시오. 이것은 두 목록 각각에서 가장 낮은 수치입니다. 또한, 각 축에 표시된 증가를 확인하실 수 있습니다(예: 2.5, 3.0, 3.5 등, y축에). 41페이지의 “Customizing Graphs”에서 그래프의 해당 측면을 제어하는 방법과 축 레이블 및 그래프 제목을 추가하는 방법을 배우게 됩니다.

X축에 해당하는 값은 목록의 각 값의 위치로 채워집니다. 즉, 첫 번째 온도 값인 53.9는 리스트의 0 위치에 있기 때문에 해당 x축 값이 0이 됩니다(리스트 위치는 1이 아니라 0부터 계산한다는 점을 기억하십시오). 그 결과, x축에 표시된 숫자는 0부터 12까지의 정수이며, 이는 우리가 온도 데이터를 보유하고 있는 13년과 대응한다고 생각할 수 있습니다.

또한 y축에 있는 숫자는 부동소수점 숫자이며 (우리가 플롯을 요청했기 때문에) x축에 있는 숫자는 정수임을 확인할 수 있습니다. Matplotlib은 두 가지 모두를 처리할 수 있습니다.

해당 연도를 표시하지 않고 온도를 플롯하는 것은 연도 간의 변동을 시각화하는 빠르고 쉬운 방법입니다. 하지만 이 그래프를 누군가에게 제시하려고 계획하고 있다면, 각 온도가 어느 연도에 해당하는지 표시하여 더 명확하게 만들고 싶을 것입니다. 우리는 연도를 포함한 다른 목록을 만든 다음 plot() 함수를 호출함으로써 쉽게 이를 수행할 수 있습니다:

우리는 1장에서 배운 range() 함수를 사용하여 2000년부터 2012년까지를 지정합니다. 이제 x축에 연도가 표시됩니다 (그림 2‐7을 참조하십시오).

플롯 함수를 한 번에 세 쌍 모두 호출하는 대신, 각 쌍당 한 번씩 세 번 호출할 수도 있습니다:

```python
plot(months, nyc_temp_2000) #[<matplotlib.lines.Line2D object at 0x7f1e51351810>]
plot(months, nyc_temp_2006) #[<matplotlib.lines.Line2D object at 0x7f1e5ae8e390>]
plot(months, nyc_temp_2012) #[<matplotlib.lines.Line2D object at 0x7f1e5136ccd0>] 
show()
```

Matplotlib은 아직 표시되지 않은 플롯을 추적합니다. 따라서 plot()을 세 번 모두 호출한 후에 show()를 호출하기만 기다리면, 모든 플롯이 동일한 그래프에 표시됩니다.

하지만 우리는 어떤 색이 어느 연도에 해당하는지 전혀 알지 못하기 때문에 문제가 있습니다. 이를 해결하려면 legend() 함수를 사용할 수 있으며, 이를 통해 그래프에 전설을 추가할 수 있습니다. 범례는 그래프의 다양한 부분이 의미하는 바를 식별하는 작은 표시 상자입니다. 여기에서는 각 색상 선이 의미하는 연도를 표시하기 위해 범례를 사용할 것입니다. 레전드를 추가하려면, 앞서와 같이 plot() 함수를 먼저 호출하십시오:

그런 다음, pylab 모듈에서 legend() 함수를 가져와서 다음과 같이 호출하십시오:

파이플로트를 이용한 플롯 그리기

Pylab 모듈은 지금까지 해오던 것처럼 IDLE 쉘과 같은 인터랙티브 쉘에서 플롯을 생성하는 데 유용합니다. 하지만 IDLE 셸 외부에서 matplotlib을 사용할 경우—예를 들어 더 큰 프로그램의 일부로—pyplot 모듈이 더 효율적입니다. 걱정하지 마세요—Pylab을 사용할 때 배운 모든 방법은 Pyplot에서도 동일하게 작동합니다.

다음 프로그램은 파이플롯 모듈을 사용하여 이 장의 첫 번째 플롯을 재현합니다:

```python
'''   Simple plot using pyplot   '''
import matplotlib.pyplot #➊
def create_graph(): #➋
    x_numbers = [1, 2, 3]
    y_numbers = [2, 4, 6]
    matplotlib.pyplot.plot(x_numbers,y_numbers)
    matplotlib.pyplot.show()
if __name__ == '__main__':
    create_graph()
```

먼저, import matplotlib .pyplot ➊ 문을 사용하여 pyplot 모듈을 가져옵니다. 이는 우리가 matplotlib 패키지에서 전체 pyplot 모듈을 가져오고 있음을 의미합니다. 이 모듈에 정의된 함수나 클래스 정의를 참조하려면, 사용하려는 함수 또는 클래스인 item에서 matplotlib.pyplot.item 구문을 사용해야 합니다.

뉴턴의 보편적 중력 법칙

뉴턴의 보편 중력 법칙에 따르면, 질량 m1인 물체가 질량 m2인 다른 물체를 힘 F의 양으로 끌어당깁니다.

여기서 r은 두 물체 사이의 거리이며 G는 중력 상수입니다. 우리는 두 물체 사이의 거리가 증가함에 따라 힘이 어떻게 작용하는지 보고 싶습니다.

두 물체의 질량을 살펴보겠습니다: 첫 번째 물체의 질량(m1)은 0.5 kg이고, 두 번째 물체(m2)의 질량은 1.5 kg입니다. 중력 상수의 값은 6.674 × 10–11 N m² kg–2입니다. 이제 우리는 이 두 물체 사이의 중력을 100 m, 150 m, 200 m, 250 m, 300 m 등 19개의 거리에서 1000 m까지 계산할 준비가 되었습니다. 다음 프로그램은 이러한 계산을 수행하고 그래프도 그립니다:

```python
'''   The relationship between gravitational force and   
distance between two bodies   '''

import matplotlib.pyplot as plt
# Draw the graph   
def draw_graph(x, y):
    plt.plot(x, y, marker='o')
    plt.xlabel('Distance in meters')
    plt.ylabel('Gravitational force in newtons')
    plt.title('Gravitational force and distance')
    plt.show()
def generate_F_r():
    # Generate values for r➊     
    r = range(100, 1001, 50)
    # Empty list to store the calculated values of F       
    F = []
    # Constant, G       
    G = 6.674*(10**-11) # Two masses
    m1 = 0.5
    m2 = 1.5    # Calculate force and add it to the list, F➋     
    for dist in r:
        force = G*(m1*m2)/(dist**2)
        F.append(force) # Call the draw_graph function➌     
        draw_graph(r, F)
        if __name__=='__main__':
            generate_F_r()
```

generate_F_r() 함수는 위 프로그램에서 대부분의 작업을 수행합니다. ➊에서, 우리는 range() 함수를 사용하여 거리의 서로 다른 값을 가진 r이라는 레이블의 리스트를 만들고, 단계값을 50으로 사용합니다. 최종 값은 1001로 지정되었습니다. 이는 1000도 포함하고자 하기 때문입니다. 그런 다음 우리는 빈 목록(F)을 생성하여 각 거리마다 해당하는 중력을 저장합니다. 다음으로, 중력 상수(G)와 두 질량(m1 및 m2)을 참조하는 라벨을 생성합니다. For 루프 ➋를 사용하여, 우리는 거리 목록(r)의 각 값에서 힘을 계산합니다. 우리는 계산된 힘을 가리키기 위해 레이블(force)을 사용하고 이를 목록(F)에 추가합니다. 마지막으로, 우리는 거리 목록과 계산된 힘 목록을 사용하여 ➌에서 함수 draw_graph()를 호출합니다. 그래프의 x축은 힘을 표시하고, y축은 거리를 표시합니다. 그래프는 그림 2‐12에 표시되어 있습니다.

거리(r)가 증가함에 따라 중력은 감소합니다. 이러한 관계에서는 중력이 두 물체 사이의 거리와 반비례한다고 말합니다. 또한, 두 변수 중 하나의 값이 변할 때, 다른 변수는 반드시 같은 비율로 변하는 것은 아닙니다. 우리는 이를 비선형 관계라고 부릅니다. 그 결과, 우리는 그래프에 직선 대신 곡선이 됩니다.

투사체 운동

이제 일상 생활에서 익숙하게 될 무언가를 그래프로 그려 보겠습니다. 공을 필드에 던지면, 그림 2‐13에 표시된 것과 같은 궤적을 따릅니다.


그림 2‐13: 각도(θ)와 속도(U)를 가진 점 A에 던져진 공의 움직임과 점 B에서 지면에 닿는 동작


그림에서 공은 A 지점에서 던져져 B 지점에 착지합니다. 이러한 종류의 움직임은 발사체 움직임이라고 합니다. 우리의 목표는 여기서 투사체 운동 방정식을 사용하여 물체의 궤적을 그래프로 표시하고, 던진 지점부터 다시 땅에 닿을 때까지 공의 위치를 표시하는 것입니다.

공을 던질 때 초기 속도가 있으며, 그 속도의 방향이 지면과 일정한 각도를 형성합니다. 초기 속도 u와 그것이 지면 θ(θ)와 이루는 각도를 그림 2‐13에 표시된 대로 호출합시다. 공은 두 개의 속도 성분을 가지고 있습니다: 하나는 x 방향으로, ux = u cosθ 로 계산되고, 다른 하나는 y 방향으로, uy = u sinθ 로 계산됩니다.

공이 움직이면 속도가 변하고, 우리는 그 변한 속도를 v를 사용하여 표현합니다: 수평 성분은 vx이고 수직 성분은 vy입니다. 단순화를 위해, 물체의 움직임 동안 수평 성분(vx)이 변하지 않는다고 가정하고, 수직 성분(vy)은 중력의 힘에 의해 감소한다고 가정합니다. vy = uy – gt 방정식에 따라. 이 방정식에서 g는 중력 가속도이며 t는 속도가 측정되는 시간입니다. Uy = u sinθ이므로, 우리는 대체하여 얻을 수 있습니다.

Vy = u sinθ – gt.

속도의 수평 성분이 일정하게 유지되기 때문에, 이동한 수평 거리(Sx)는 Sx = u(cosθ)t 로 주어집니다. 하지만 속도의 수직 성분은 변하고, 이동한 수직 거리는 공식에 의해 주어집니다.


다시 말해, Sx와 Sy는 공이 비행하는 동안 주어진 시점의 x와 y 좌표를 우리에게 제공합니다. 우리는 궤적을 그리는 프로그램을 작성할 때 이 방정식들을 사용할 것입니다. 이러한 방정식을 사용할 때, 시간(t)은 초 단위로 표현되고, 속도는 m/s 단위로 표현되며, 투영각(θ)은 도 단위로 표현되고, 중력 가속도(g)는 m/s2 단위로 표현됩니다.

하지만 프로그램을 작성하기 전에, 공이 땅에 닿기 전에 비행 중일 수 있는 시간을 확인하여, 우리 프로그램이 공의 궤적을 그리는 것을 언제 멈춰야 하는지 알아야 합니다. 이를 위해 우리는 먼저 공이 최고점에 도달하는 데 걸리는 시간을 찾겠습니다. 공은 속도의 수직 성분(vy)이 0일 때 최고점에 도달합니다. 이는 vy = u sin θ – gt = 0일 때입니다. 그래서 우리는 공식을 사용하여 t 값을 찾고 있습니다.

이번 시간을 t_peak이라고 부르겠습니다. 공이 최고점에 도달한 후, 공은 또 다른 t_peak 초 동안 공중에 떠 있는 뒤 땅에 충돌하므로, 공의 총 비행 시간(t_flight)은

초기 속도(u)가 5 m/s이고 각도(θ) 45도인 공을 잡아 보겠습니다. 비행 총 시간을 계산하기 위해, 우리는 위에서 본 식에 u = 5, θ = 45, g = 9.8을 대입합니다:

이 경우, 공의 비행 시간은 0.72154초(소수점 다섯 자리까지 반올림)입니다. 공은 이 기간 동안 공중에 있을 것이므로, 궤적을 그리기 위해 우리는 이 기간 동안 일정한 간격으로 x와 y 좌표를 계산합니다. 좌표를 얼마나 자주 계산해야 합니까? 가능하면 가능한 한 자주. 이 장에서는 0.001초마다 좌표를 계산합니다.


균등 간격 부동 소수점 숫자 생성 
우리는 range() 함수를 사용하여 균등 간격의 정수를 생성했습니다—즉, 1에서 10 사이의 정수를 1로 구분한 리스트를 원한다면 range(1, 10)을 사용합니다. 다른 단계 값을 원한다면, 이를 range 'function을 세 번째 인수로 지정할 수 있습니다. 안타깝게도 부동소수점 숫자에 대한 내장 함수는 없습니다. 예를 들어, 0부터 0.72까지의 두 연속된 숫자를 0.001로 구분하여 목록을 만들 수 있는 함수는 없습니다. 우리는 다음과 같이 while 루프를 사용하여 이를 위한 자체 함수를 만들 수 있습니다:

```python
'''   
Generate equally spaced floating point   
numbers between two given values   
'''
def frange(start, final, increment):
    numbers = []
    while start < final: #➊
        numbers.append(start)#➋           
        start = start + increment
        return numbers
```

우리는 frange()(“floating point” range) 함수를 정의했으며, 이 함수는 세 가지 매개변수를 받습니다: start와 final은 숫자 범위의 시작점과 끝점을 가리키고, increment는 연속된 두 숫자의 차이를 나타냅니다. 우리는 ➊에 while 루프를 초기화하고, start가 가리키는 숫자가 final 값보다 작을 때까지 실행을 계속합니다. 우리는 start가 가리키는 숫자를 목록 번호 ➋에 저장한 다음, 루프의 각 반복마다 증가분으로 입력한 값을 추가합니다. 마지막으로, 목록 번호를 반환합니다.

다음에 설명할 궤적 그리기 프로그램에서 동일한 간격으로 시간 순간을 생성하기 위해 이 함수를 사용할 것입니다.



궤적 그리기

다음 프로그램은 일정한 속도와 각도로 던져진 공의 궤적을 그리며, 이 두 요소는 프로그램에 입력으로 제공됩니다:

```python
'''   Draw the trajectory of a body in projectile motion   '''
from matplotlib import pyplot as plt
import math
def draw_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('x-coordinate')
    plt.ylabel('y-coordinate')
    plt.title('Projectile motion of a ball')
def frange(start, final, interval):
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + interval
        return numbers
def draw_trajectory(u, theta):

    theta = math.radians(theta) #➊  
    g = 9.8
    # Time of flight
    t_flight = 2*u*math.sin(theta)/g    # ➋         
    # Find time intervals       
    intervals = frange(0, t_flight, 0.001)
    # List of x and y coordinates       
    x = []
    y = []
    for t in intervals:     #➌              
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t - 0.5*g*t*t)
        draw_graph(x, y)
        if __name__ == '__main__':
            try: #➍                
                u = float(input('Enter the initial velocity (m/s): '))
                theta = float(input('Enter the angle of projection (degrees): '))
            except ValueError:
                print('You entered an invalid input')
        else:
            draw_trajectory(u, theta)
            plt.show()
```

이 프로그램에서는 표준 라이브러리의 수학 모듈에 정의된 함수 radians(), cos(), sin()를 사용하므로 해당 모듈을 처음에 가져옵니다. Draw_trajectory() 함수는 공이 던져지는 속도와 각도에 해당하는 두 개의 인수 u와 theta를 받습니다. 수학 모듈의 사인과 코사인 함수는 각도가 라디안으로 제공되기를 기대하므로, ➊에서 우리는 각도(세타)를 도에서 라디안으로 변환하여 math.radians() 함수를 사용합니다. 다음으로, 중력에 의한 가속도 값인 9.8 m/s²를 나타내는 라벨 (g)을 생성합니다. ➋에서 우리는 비행 시간을 계산한 다음, start, final, increment 값을 각각 0, t_flight, 0.001로 설정하여 frange() 함수를 호출합니다. 그런 다음 우리는 각 시간 순간에서 궤적의 x와 y 좌표를 계산하고 이를 두 개의 별도 목록인 x와 y ➌에 저장합니다. 이러한 좌표를 계산하기 위해, 우리는 앞서 논의한 거리 Sx와 Sy에 대한 공식을 사용합니다.

마지막으로, 우리는 x-좌표와 y-좌표를 사용하여 draw_graph() 함수를 호출하여 궤적을 그립니다. Draw_graph() 함수는 show() 함수를 호출하지 않는다는 점에 유의하십시오(다음 프로그램에서 그 이유를 확인하겠습니다). 우리는 사용자가 잘못된 입력을 할 경우 오류 메시지를 보고하기 위해 try...except 블록 ➍을 사용합니다. 이 프로그램에 대한 유효한 입력은 정수 또는 부동소수점 숫자입니다. 프로그램을 실행하면 이러한 값을 입력으로 요청한 뒤 궤적을 그립니다(그림 2‐14 참조):

#5: 피보나치 수열과 황금비 사이의 관계 탐구 피보나치 수열 (1, 1, 2, 3, 5, ...)은 수열의 i번째 수가 이전 두 수의 합인 숫자들의 연속이며, 즉 (i – 2)와 (i – 1) 위치의 숫자를 의미합니다. 이 시리즈의 연속된 숫자들은 흥미로운 관계를 보여줍니다. 급수에서 항의 수를 늘릴 때, 연속된 숫자 쌍들의 비율은 거의 서로 같습니다. 이 값은 황금비라고 불리는 특수한 수에 접근합니다. 수치적으로 황금비는 숫자 1.618033988 ...이며, 이는 음악, 건축 및 자연 분야에서 광범위하게 연구되어 왔습니다.



이 도전 과제에서는 예를 들어 100개의 숫자에 대한 연속된 피보나치 수들 간의 비율을 그래프에 표시하는 프로그램을 작성하여, 그 값이 황금비에 근접함을 증명하도록 하십시오. 다음 함수를 찾을 수 있습니다. 이 함수는 처음 n개의 피보나치 숫자 목록을 반환하며, 귀하의 해답을 구현하는 데 유용합니다.

```python
def fibo(n):
    if n == 1:
        return [1]
    if n == 2:
        return [1,1]

    # n > 2    
    a = 1
    b = 1
    # First two members of the series    
    series = [a,b]
    for i in range(n-2): 
        c = a + b 
        series.append(c) 
        a = b 
        b = c 
    return series
```

##### 평균 찾기

평균은 숫자 집합을 요약하는 일반적이고 직관적인 방법입니다. 일상적인 사용에서는 단순히 “평균”이라고 부를 수 있지만, 곧 보게 될 다른 종류의 평균도 존재합니다. 샘플 숫자 집합을 사용하여 평균을 계산해 보겠습니다.
<br><br>
예를 들어, 지난 12일 동안 기부를 받고 있는 학교 자선 단체가 있다고 가정해 보겠습니다(이를 '기간 A'라고 부르겠습니다). 그 기간 동안, 다음 12개의 숫자는 매일 받은 기부금 총액을 나타냅니다: 100, 60, 70, 900, 100, 200, 500, 500, 503, 600, 1000, 그리고 1200. 우리는 이 총합을 합산한 뒤 그 합을 일수로 나누어 평균을 계산할 수 있습니다. 이 경우, 숫자들의 합은 5733입니다. 이 수치를 12(일수)로 나누면 477.75가 되며, 이는 하루 평균 기부액입니다. 이 숫자는 특정 날짜에 기부된 금액을 대략적으로 파악할 수 있게 해줍니다.

잠시 후, 우리는 숫자 집합의 평균을 계산하고 출력하는 프로그램을 작성하겠습니다. 방금 보았듯이, 평균을 계산하려면 숫자 목록의 합을 그 목록의 항목 수로 나누어야 합니다. 두 개의 파이썬 함수를 살펴보겠습니다: 이 두 연산을 모두 매우 쉽게 수행할 수 있습니다: sum()와 len().

sum() 함수를 숫자 목록에 사용하면, 목록에 있는 모든 숫자를 합산하여 결과를 반환합니다:

##### 평균을 계산하기  
''' python
def calculate_mean(numbers):
    s = sum(numbers)    #➊ 
    N = len(numbers)    #➋  평균을 계산합니다
    mean = s/N              #➌ 평균 
    return mean 
    if __name__ == '__main__':
        donations = [100, 60, 70, 900, 100, 200, 500, 500, 503, 600, 1000, 1200]    #➍ 
        mean = calculate_mean(donations)    #➎
        N = len(donations)          
        print('지난 {0}일 동안의 평균 기부는 {mean} 입니다')         #➏ 
'''

먼저, 우리는 numbers라는 인수 numbers를 받아들이는 함수인 calculate_mean()를 정의합니다. ➊에서 우리는 sum() 함수를 사용하여 목록의 숫자를 더하고, 총합을 나타내는 레이블 s를 생성합니다. 마찬가지로, ➋에서 우리는 len() 함수를 사용하여 목록의 길이를 구하고, 이를 참조하기 위해 레이블 N을 생성합니다. 그렇다면, ➌에서 보시다시피, 우리는 합(s)을 멤버 수(N)로 간단히 나누어 평균을 계산합니다. ➍에서는 앞서 나열된 기부금의 가치를 사용하여 기부 목록을 작성합니다. 그런 다음 calculate_mean() 함수를 호출하고, 이 목록을 ➎에 인수로 전달합니다. 마지막으로, ➏에서 계산된 평균을 출력합니다.
프로그램을 실행하면 다음과 같은 내용이 나타납니다:

```python
aMoney = [100, 60, 70, 900, 100, 200, 500, 500, 503, 600, 1000, 1200]
mMoney = sum(aMoney) / len(aMoney)
print(mMoney)
```

```python
def calculate_mean(numbers):
    s = sum(numbers)    #➊ 
    N = len(numbers)    #➋  평균을 계산합니다
    mean = s/N              #➌ 평균 
    return mean 
if __name__ == '__main__':
    donations = [100, 60, 70, 900, 100, 200, 500, 500, 503, 600, 1000, 1200]    #➍ 
    mean = calculate_mean(donations)    #➎
    print (mean)
    N = len(donations)          
    print('지난 {0} 일 동안의 평균 기부는 {1} 입니다'.format(N, mean))         #➏ 
```

Calculate_mean() 함수는 모든 목록의 합과 길이를 계산하므로, 다른 숫자 집합의 평균을 계산하는 데도 이를 재사용할 수 있습니다.

우리는 하루 평균 기부액이 477.75임을 계산했습니다. 주목할 만한 점은 처음 며칠간의 기부금이 우리가 계산한 평균 기부금보다 훨씬 낮으며, 마지막 며칠간의 기부금은 훨씬 높았다는 것입니다. 평균은 데이터를 요약할 수 있는 한 가지 방법을 제공하지만, 전체적인 그림은 제공하지 않습니다. 그러나 평균과 비교할 때 데이터에 대해 더 많은 정보를 알려줄 수 있는 다른 통계 측정값도 있습니다.

##### 중앙값 찾기

숫자 집합의 중앙값은 또 다른 종류의 평균입니다. 중앙값을 찾기 위해 우리는 숫자를 오름차순으로 정렬합니다. 숫자 목록의 길이가 홀수인 경우, 목록 가운데에 있는 숫자는 중앙값입니다. 숫자 목록의 길이가 짝수인 경우, 두 중간 숫자의 평균을 취하여 중앙값을 구합니다. 이전 기부 목록의 중앙값을 찾아봅시다: 100, 60, 70, 900, 100, 200, 500, 500, 503, 600, 1000, 그리고 1200.

가장 작은 것부터 가장 큰 것까지 정렬한 후, 숫자 목록은 60, 70, 100, 100, 200, 500, 500, 503, 600, 900, 1000, 그리고 1200이 됩니다. 우리는 목록(12)에 짝수 개의 항목이 있으므로 중앙값을 구하려면 두 중간 숫자의 평균을 구해야 합니다. 이 경우, 중간 숫자는 여섯 번째와 일곱 번째 숫자인 500과 500이며, 이 두 숫자의 평균은 (500 + 500)/2이며, 이는 500이 됩니다. 그것은 중앙값이 500임을 의미합니다.

이제 이 예를 들어 가정해 보겠습니다—13일째에 또 다른 기부 총액이 있다고 가정하면 목록은 다음과 같습니다: 100, 60, 70, 900, 100, 200, 500, 500, 503, 600, 1000, 1200, 그리고 800.

다시 한 번, 우리는 목록을 정렬해야 하며, 그 목록은 60, 70, 100, 100, 200, 500, 500, 503, 600, 800, 900, 1000, 그리고 1200이 됩니다. 이 목록에는 13개의 숫자(홀수)가 있으므로, 이 목록의 중앙값은 단순히 가운데 숫자입니다. 이 경우, 그것은 일곱 번째 숫자이며, 500입니다.

숫자 목록의 중앙값을 찾는 프로그램을 작성하기 전에, 어느 경우든 리스트의 중간 요소를 자동으로 계산할 수 있는 방법을 생각해 봅시다. 리스트(N)의 길이가 홀수인 경우, 중간 숫자는 위치(N + 1)/2에 있는 숫자입니다. N이 짝수라면, 두 중간 요소는 N/2와 (N/2) + 1입니다. 이 절의 첫 번째 예에서는 N = 12이며, 따라서 두 중간 요소는 12/2(여섯 번째)와 12/2 + 1(여섯 번째) 요소였습니다. 두 번째 예에서 N = 13, 따라서 일곱 번째 원소인 (N + 1)/2가 중간이었습니다.

중앙값을 계산하는 함수를 작성하려면, 목록을 오름차순으로 정렬해야 합니다. 다행히도, sort() 메서드는 바로 그 일을 수행합니다:

```python
samplelist=[4,1,3]
samplelist.sort()
print(samplelist)
```

이제 우리는 숫자 목록의 중앙값을 찾는 다음 프로그램을 작성할 수 있습니다:

```python
''' 중앙값 계산 중위값을 계산합니다 ''' 
def calculate_median(numbers): 
    N = len(numbers)    #➊
    numbers.sort()      #➋  
    # N이 짝수인 경우 중앙값을 찾습니다 
    if N % 2 == 0: 
        # N이 짝수인 경우 
        m1 = N/2 
        m2 = (N/2) + 1 
        # 정수로 변환하여 위치를 맞춥니다
        m1 = int(m1) - 1    #➌ 
        m2 = int(m2) - 1
        median = (numbers[m1] + numbers[m2])/2
    else:  # N이 홀수인 경우
        m = (N+1)/2
        # 정수로 변환하여 위치와 일치합니다
        m = int(m) - 1
        median = numbers[m]
    return median

if __name__ == '__main__':
    donations = [100, 60, 70, 900, 100, 200, 500, 500, 503, 600, 1000, 1200]

    median = calculate_median(donations) 
    N = len(donations) 
    print('지난 {0}일 동안의 기부 중앙값 : {1}'.format(N, median))   

```

프로그램의 전체 구조는 평균을 계산하는 이전 프로그램과 유사합니다.

Calculate_median() 함수는 숫자 목록을 받아 중앙값을 반환합니다. ➊에서 우리는 목록의 길이를 계산하고 이를 참조하기 위해 레이블 N을 생성합니다. 다음으로, ➋에서 sort() 메서드를 사용하여 목록을 정렬합니다.

그런 다음, N이 짝수인지 확인합니다. 그렇다면, 정렬된 목록에서 N/2와 (N/2) + 1 위치에 있는 숫자인 중간 요소인 m1과 m2를 찾습니다. 다음 두 문장(➌ 및 ➍)은 m1과 m2를 두 가지 방식으로 조정합니다. 먼저, int() 함수를 사용하여 m1과 m2를 정수 형태로 변환합니다.  <span style = 'font-size:20px; color:red;'>이는 나눗셈 연산자의 결과가 정수와 동일하더라도 항상 부동소수점 숫자로 반환되기 때문입니다. </span> 예를 들어:

```python
int(6/2)
6/2
```

리스트에서 부동소수점 숫자를 인덱스로 사용할 수 없으므로, 해당 결과를 정수로 변환하기 위해 int()를 사용합니다. 우리는 또한 m1과 m2 모두에서 1을 빼는데, 이는 파이썬에서 목록의 위치가 0으로 시작하기 때문입니다. 이는 목록에서 여섯 번째와 일곱 번째 숫자를 얻기 위해 인덱스 5와 인덱스 6에 있는 숫자를 요청해야 함을 의미합니다. ➎에서 우리는 중간 위치에 있는 두 수의 평균을 사용하여 중앙값을 계산합니다.
➏에서 시작하여, 프로그램은 목록에 항목이 홀수 개수인 경우 중앙값을 찾으며, 다시 한 번 int()를 사용하고 1을 빼서 올바른 인덱스를 찾습니다. 마지막으로, 프로그램은 기부 목록의 중앙값을 계산하여 반환합니다. 프로그램을 실행하면 중앙값이 500임을 계산합니다:
‘Median donation over the last 12 days is 500.0’
보시다시피, 평균(477.75)과 중앙값(500)은 이 특정 목록에서 꽤 가깝지만, 중앙값은 약간 더 높습니다.

##### 모드 찾기 및 빈도표 만들기

숫자 집합의 평균값이나 중앙값을 찾는 대신, 가장 자주 나타나는 숫자를 찾고 싶다면 어떨까요? 이 번호는 <span style = 'font-size:20px; color:red;'> 모드 ==> 최빈값 </span> 라고 합니다. 예를 들어, 20명의 학생으로 구성된 반에서 수학 시험 점수(10점 중)를 고려하십시오: 7, 8, 9, 2, 10, 9, 9, 9, 9, 4, 5, 6, 1, 5, 6, 7, 8, 6, 1, 그리고 10. 이 목록의 모드는 수업에서 가장 흔한 점수가 어느 것인지 알려드릴 것입니다. 목록에서 9점이 가장 자주 나타나는 것을 확인할 수 있으므로, 9가 이 숫자 목록의 모드입니다. 모드를 계산하기 위한 상징적 공식은 없습니다—각 고유 숫자가 발생하는 횟수를 세고 가장 많이 발생하는 숫자를 찾기만 하면 됩니다.

모드를 계산하는 프로그램을 작성하려면, 파이썬에게 각 숫자가 목록 내에서 몇 번 나타나는지 계산하고 가장 자주 나타나는 숫자를 출력하도록 해야 합니다. 표준 라이브러리의 일부인 컬렉션 모듈의 카운터 클래스는 우리에게 이를 매우 간단하게 만들어 줍니다.

두 개 이상의 숫자가 동일한 최대 횟수로 나타나는 데이터 집합이 있다면 어떻게 됩니까? 예를 들어, 숫자 5, 5, 5, 4, 4, 4, 9, 1, 3의 목록에서 4와 5가 모두 세 번 존재합니다. 이러한 경우에는 숫자 목록에 여러 모드가 있다고 하며, 우리 프로그램은 모든 모드를 찾아 출력해야 합니다. 수정된 프로그램은 다음과 같습니다:

```python
'''   
Calculating the mode when the list of numbers may   
have multiple modes   
'''

from collections import Counter

def calculate_mode(numbers):
    c = Counter(numbers)          
    numbers_freq = c.most_common()  #➊
    print ("numbers_freq : ", numbers_freq)
    max_count = numbers_freq[0][1]  #➋
    print ("max_count : ", max_count)
    modes = []
    
    for num in numbers_freq:
        if num[1] == max_count:  #➌
            modes.append(num[0])
    print ("modes : ", modes)
    return modes

if __name__ == '__main__':
    scores = [5, 5, 5, 4, 4, 4, 9, 1, 3, 2, 1, 3, 2, 1, 3, 2]
    modes = calculate_mode(scores)
    print('The mode(s) of the list of numbers are:')
    
    for mode in modes:  #➍
        print(mode)
```

➊에서 가장 일반적인 요소만 찾는 대신, 모든 숫자와 각 요소가 나타나는 횟수를 가져옵니다. <br> 다음으로, ➋에서 최대 카운트 값을 찾습니다 — 즉, 어떤 수가 발생하는 최대 횟수입니다. <br> 그런 다음, 각 숫자에 대해 나타나는 횟수가 최대 횟수 ➌와 같은지 확인합니다. <br> 이 조건을 만족하는 각 숫자는 모드이며, 우리는 이를 리스트 모드에 추가하고 리스트를 반환합니다. <br>
➍에서, 우리는 calculate_mode() 함수에서 반환된 목록을 반복하고 각 숫자를 출력합니다.

우리는 다시 한 번 most_common() 메서드를 사용하여 주어진 숫자 집합에 대한 빈도표를 출력하겠습니다. Most_common() 메서드에 인수를 제공하지 않을 경우, 모든 숫자와 해당 메서드가 나타나는 횟수를 포함한 튜플 목록을 반환한다는 점을 기억하십시오. 우리는 이 목록에서 각 숫자와 그 빈도를 간단히 인쇄하여 빈도표를 표시할 수 있습니다.

```python
''' 컬렉션에서 가져온 숫자 목록에 대한 빈도 테이블 ''' 
from collections import Counter 
def frequency_table(numbers):
    table = Counter(numbers) #➊ 

    print('숫자\t빈도')
    for number in table.most_common(): #➋ 
        print('{0}\t{1}'.format(number[0], number[1])) 
if __name__=='__main__': 
    scores = [7,8,9,2,10,9,9,9,4,5,6,1,5,6,7,8,6,1,10] 
    frequency_table(scores)
```

##### 숫자 집합의 범위 찾기

다시 한 번, A 기간 동안의 기부 목록을 고려하십시오: 100, 60, 70, 900, 100, 200, 500, 500, 503, 600, 1000, 그리고 1200. 우리는 하루 평균 기부가 477.75임을 발견했습니다. 하지만 평균만을 살펴보면, 모든 기부금이 400에서 500 사이의 좁은 범위에 해당했는지, 혹은 그보다 훨씬 더 크게 변동했는지—예를 들어 이 경우처럼 60에서 1,200 사이인지 전혀 알 수 없습니다. 숫자 목록의 경우, 범위는 가장 높은 수와 가장 낮은 수 사이의 차이입니다. 두 개의 숫자 그룹이 정확히 동일한 평균을 가지고 있지만 범위가 크게 다를 수 있으므로, 그 범위를 알면 평균, 중앙값 및 모드만 보고 배울 수 있는 것보다 더 많은 숫자 집합에 대한 정보를 채워줍니다.

다음 프로그램은 범위를 찾습니다

```python
''' 범위 찾기 ''' 
def find_range(numbers):
    lowest = min(numbers)    #➊ 
    highest = max(numbers)   #➋ 

    # 범위 찾기 
    r = highest-lowest

    return lowest, highest, r   #➌ 

if __name__ == '__main__': 
    donations = [100, 60, 70, 900, 100, 200, 500, 500, 503, 600, 1000, 1200]
    lowest, highest, r = find_range(donations)  #➍
    print('최저: {0} 최고: {1} 범위: {2}'.format(lowest, highest, r))
```

함수 find_range()는 리스트를 매개변수로 받아 범위를 찾습니다. <br> 먼저, ➊와 ➋에서 min() 및 max() 함수를 사용하여 가장 낮은 수와 가장 높은 값을 계산합니다. 함수 이름이 나타내듯이, 그들은 숫자 목록에서 최소값과 최대값을 찾습니다.

그런 다음 우리는 가장 높은 수와 가장 낮은 수 사이의 차이를 취하고, 라벨 r을 사용하여 그 차이를 나타내는 범위를 계산합니다. <br> ➌에서 우리는 세 개의 숫자를 모두 반환합니다—가장 낮은 숫자, 가장 높은 숫자, 그리고 범위. 이번 책에서 함수에서 여러 값을 반환하는 것은 이번이 처음입니다 — 하나의 값만 반환하는 대신, 이 함수는 세 개를 반환합니다. <br> ➍에서 우리는 find_range() 함수에서 반환되는 세 값을 받기 위해 세 개의 라벨을 사용합니다. <br> 마지막으로, 우리는 값을 출력합니다. 프로그램을 실행하면 다음과 같은 출력이 나타납니다:

최저: 60 최고: 1200 범위: 1140

이는 하루 총 기부금이 1140 범위로 <span style = 'font-size:20px; color:green;'>꽤 넓게 분포되어 있음을 알려줍니다. </span> 이는 일일 총액이 60에서 1,200까지 있었기 때문입니다.

##### 분산 및 표준편차 찾기

범위는 숫자 집합에서 두 극단 사이의 차이를 알려 주지만, 모든 개별 숫자가 평균과 어떻게 다른지 더 알고 싶다면 어떨까요? 그들은 모두 평균에 가깝게 군집된 것이 비슷했습니까, 아니면 모두 다른, 극에 더 가깝게 했습니까? 분산에 관한 두 가지 관련된 측정 지표가 있으며, 이는 분산과 표준편차와 같은 숫자 목록에 대해 더 자세히 알려줍니다. 이 중 하나를 계산하려면 먼저 평균으로부터 각 숫자의 차이를 찾아야 합니다. 분산은 그 차이들의 제곱들의 평균입니다. 높은 분산은 값이 평균에서 멀리 떨어져 있음을 의미하고, 낮은 분산은 값이 평균에 가깝게 군집화되어 있음을 의미합니다. 우리는 공식을 사용하여 분산을 계산합니다.

공식에서 xi는 개별 숫자(이 경우 일일 총 기부액)를 의미하고, xmean은 해당 숫자들의 평균(평균 일일 기부액)을 의미하며, n은 목록의 값 수(기부가 접수된 일수)를 의미합니다. 목록의 각 값에 대해 해당 숫자와 평균의 차이를 곱합니다. 그런 다음, 우리는 모든 제곱 차이를 더하고, 마지막으로 전체 합을 n으로 나누어 분산을 구합니다.

표준편차도 계산하고 싶다면, 우리가 해야 할 일은 분산의 제곱근을 취하는 것뿐입니다. 평균보다 표준편차가 1표준편차 이내인 값은 비교적 일반적인 것으로 생각할 수 있는 반면, 평균으로부터 표준편차가 3개 이상 떨어진 값은 훨씬 더 비전형적인 것으로 간주될 수 있습니다—우리는 이러한 값을 이상치라고 부릅니다.

왜 우리는 분산의 두 척도인 분산과 표준편차를 가지고 있습니까? 요약하면, 두 가지 조치는 서로 다른 상황에서 유용합니다. 우리가 분산을 계산하는 데 사용한 공식으로 돌아가면, 분산이 제곱 단위로 표현된다는 것을 알 수 있습니다. 이는 제곱의 평균이기 때문입니다. 일부 수학 공식에서는 표준편차를 찾기 위해 제곱근을 사용하는 것보다 그 제곱 단위를 사용하는 것이 더 좋습니다. 반면에, 표준 편차는 인구 데이터와 동일한 단위로 표시됩니다.

예를 들어, 우리 기부 목록의 분산을 계산한다면(잠시 후에 계산하겠습니다), 결과는 달러 제곱으로 표시되며, 이는 크게 의미가 없습니다. 한편, 표준 편차는 단순히 달러로 표시되며, 각 기부금과 동일한 단위입니다.

다음 프로그램은 숫자 목록에 대한 분산과 표준 편차를 찾습니다:



```python
''' 숫자 목록의 분산과 표준편차를 구하십시오 ''' 
def calculate_mean(numbers): 
    s = sum(numbers) 
    N = len(numbers) 
    # 평균을 계산합니다 평균 = s/N 반환 평균 
    mean = s/N
    
    return mean

def find_differences(numbers): 
    # Find the mean
    mean = calculate_mean(numbers)
    # Find the differences from the mean
    diff = []
    for num in numbers:
        diff.append(num - mean)
    return diff

def calculate_variance(numbers):
    # Find the list of differences
    diff = find_differences(numbers)    #➊
    # Find the squared differences
    squared_diff = []
    for d in diff:      #➋
        squared_diff.append(d**2)
    # Find the variance
    sum_squared_diff = sum(squared_diff)
    variance = sum_squared_diff / len(numbers)  #➌
    return variance

if __name__ == '__main__':
    donations = [100, 60, 70, 900, 100, 200, 500, 500, 503, 600, 1000, 1200]
    variance = calculate_variance(donations) 
    print('숫자 목록의 분산은 {0}'.format(variance))
    
    std = variance**0.5     #➍
    print('숫자 목록의 표준편차는 {0}'.format(std))
```

함수 calculate_variance()는 전달된 숫자 목록의 분산을 계산합니다. 먼저, ➊에서 find_differences() 함수를 호출하여 평균으로부터 각 숫자의 차이를 계산합니다. find_differences() 함수는 각 기부와 평균값 사이의 차이를 목록 형태로 반환합니다. 이 함수에서는 앞서 작성한 calculate_mean() 함수를 사용하여 평균 기부를 찾습니다. 그런 다음, ➋에서 시작하여 이러한 차이들의 제곱을 계산하고 squared_diff라는 레이블이 붙은 목록에 저장합니다. 다음으로, 우리는 sum() 함수를 사용하여 제곱 차이의 합을 찾고, 마지막으로 ➌에서 분산을 계산합니다. ➍에서 우리는 분산의 제곱근을 취하여 표준 편차를 계산합니다.

두 데이터 세트 간의 상관관계 계산

이 섹션에서는 두 집합의 관계의 성질과 강도를 알려주는 통계적 측정값을 계산하는 방법을 배우게 됩니다: 피어슨 상관계수이며, 저는 이를 단순히 상관계수라고 부르겠습니다. 이 계수는 선형 관계의 강도를 측정한다는 점에 유의하십시오. 두 집합이 비선형 관계를 가질 때 계수를 찾기 위해 다른 측정값을 사용해야 합니다(여기서는 논의하지 않을 것입니다). 계수는 양수이거나 음수일 수 있으며, 그 크기는 –1에서 1(포함) 사이일 수 있습니다.

상관계수가 0이면 두 양 사이에 선형 상관관계가 없음을 나타냅니다. (이는 두 양이 서로 독립적이라는 의미가 아니라는 점에 유의하십시오.) 예를 들어, 그들 사이에 여전히 비선형 관계가 존재할 수 있습니다. 계수가 1이거나 1에 가깝다는 것은 강한 양의 선형 상관관계가 있음을 나타내며, 정확히 1인 계수를 완전한 양의 상관관계라고 합니다. 마찬가지로, 상관계수가 –1이거나 –1에 근접하면 강한 음의 상관관계를 나타내며, 여기서 1은 완전한 음의 상관관계를 의미합니다.

##### 상관관계와 인과관계

통계학에서는 종종 “상관관계가 인과관계를 의미하지 않는다”는 진술을 접하게 됩니다. 이는 두 관측 집합이 서로 강하게 상관관계가 있더라도, 한 변수가 다른 변수를 원인으로 만든다는 의미는 아니라는 점을 상기시키는 것입니다. 두 변수가 강하게 상관관계가 있을 때, 때때로 두 변수에 영향을 미치는 세 번째 요인이 존재하여 상관관계를 설명합니다. 전형적인 예는 아이스크림 판매와 범죄율 사이의 상관관계입니다—일반적인 도시에서 이 두 변수를 모두 추적하면 상관관계를 찾을 가능성이 높지만, 이것이 아이스크림 판매가 범죄를 일으킨다는 의미는 아닙니다(또는 그 반대도 마찬가지입니다). 아이스크림 판매와 범죄는 여름 동안 날씨가 더워질수록 상승하기 때문에 상관관계가 있습니다. 물론, 이것이 더운 날씨가 직접적으로 범죄를 증가시키는 것을 의미하는 것은 아니며, 그 연관성 뒤에도 더 복잡한 원인이 있습니다.

상관계수 계산

상관계수는 공식을 사용하여 계산됩니다.

위 공식에서 n은 각 숫자 집합에 존재하는 값의 총 개수이며(집합의 길이는 동일해야 합니다). 두 개의 숫자 집합은 x와 y로 표시됩니다(어디를 어느 것으로 표시하든 상관없습니다). 다른 용어는 다음과 같이 설명됩니다:

Σxy     X와 y 두 집합의 개별 원소들의 곱 합
Σx      집합 x에 있는 수들의 합
Σy      집합 y에 있는 숫자들의 합
(Σx)2   집합 x에 있는 수들의 합의 제곱
(Σy)2   집합 y에 있는 수들의 합의 제곱
Σx2     집합 x에 있는 숫자들의 제곱합
Σy2     집합 y에 있는 숫자들의 제곱합

이러한 항들을 계산한 후에는 앞선 공식에 따라 이를 결합하여 상관계수를 찾을 수 있습니다. 작은 목록의 경우, 큰 노력 없이 손으로 할 수 있지만, 각 숫자 집합의 크기가 커짐에 따라 확실히 복잡해집니다.

잠시 후, 우리를 위해 상관계수를 correlation coefficient 계산하는 프로그램을 작성하겠습니다. 이 프로그램에서는 zip() 함수를 사용할 것이며, 이는 두 숫자 집합의 곱합을 계산하는 데 도움이 됩니다. 다음은 zip() 함수가 어떻게 작동하는지에 대한 예시입니다:

```python
simple_list1 = [1, 2, 3]
simple_list2 = [4, 5, 6]
for x, y in zip(simple_list1, simple_list2):
    print(x, y)
```

Zip() 함수는 x와 y에 해당하는 요소의 쌍을 반환하며, 이를 루프에서 사용하여 이전 코드에 표시된 인쇄와 같은 다른 작업을 수행할 수 있습니다. 두 리스트의 길이가 동일하지 않으면, 더 작은 리스트의 모든 요소가 읽히면 함수가 종료됩니다.

이제 우리는 우리를 위해 상관계수를 계산해 줄 프로그램을 작성할 준비가 되었습니다:

```python
def find_corr_x_y(x,y): 
    n = len(x)  #label n 을 생성
    
    # products의 합을 찾습니다 
    prod = []
    for xi, yi in zip(x, y):    #➊
        prod.append(xi*yi)

    print ('prod: ', prod) 

    sum_prod_x_y = sum(prod)    #➋
    sum_x = sum(x)              #➌
    sum_y = sum(y)              #➍
    squared_sum_x = sum_x**2
    squared_sum_y = sum_y**2

    print ('squared_sum_x: ',squared_sum_x)
    print ('squared_sum_y: ',squared_sum_y)

    x_square = []
    for xi in x:        #➎
        x_square.append(xi**2)
    # 합을 찾습니다 
    x_square_sum = sum(x_square)
  
    y_square=[]
    for yi in y:
        y_square.append(yi**2)
    # 합을 찾습니다 
    y_square_sum = sum(y_square)
    # 상관계수를 계산 위한 공식 사용 
    numerator = n*sum_prod_x_y - sum_x*sum_y    #➏
    denominator_term1 = n*x_square_sum - squared_sum_x
    denominator_term2 = n*y_square_sum - squared_sum_y
    denominator = (denominator_term1*denominator_term2)**0.5    #➐
    correlation = numerator/denominator             #➑
    print ('상관계수: {0}'.format(correlation))
    return correlation
```

find_corr_x_y() 함수는 두 개의 인수인 x와 y를 받으며, 이는 우리가 상관관계를 계산하고자 하는 두 개의 숫자 집합입니다. <br> 이 함수의 시작 부분에서 우리는 목록의 길이를 찾아 이를 가리키는 레이블 n을 생성합니다. 다음으로, <br>  ➊에서는 zip() 함수를 사용하여 각 리스트의 해당 값들의 곱을 계산하는 for 루프가 있습니다(각 리스트의 첫 번째 항목을 곱하고, 그 다음 두 번째 항목을 곱하는 등). 우리는 append() 메서드를 사용하여 이러한 제품들을 “prod”라는 레이블이 붙은 목록에 추가합니다. <br>
➋에서 우리는 prod에 저장된 곱들의 합을 sum() 함수를 사용하여 계산합니다. <br> ➌와 ➍에 있는 문장에서는 각각 x와 y의 숫자 합을 계산합니다(다시 한 번 sum() 함수를 사용하여). 그런 다음, 우리는 x와 y에 있는 원소들의 합의 제곱을 계산하고, 각각 이를 참조하기 위해 label squared_sum_x와 squared_sum_y를 생성합니다.  <br>

루프에서 ➎에서 시작하여, 우리는 x의 각 원소들의 제곱을 계산하고 그 제곱들의 합을 구합니다. 그럼, y의 원소에 대해서도 동일하게 합니다. 우리는 이제 상관관계를 계산하는 데 필요한 모든 항을 가지고 있으며, 이를 ➏, ➐, ➑에 있는 문장에서 수행합니다. 마지막으로, 우리는 상관관계를 반환합니다. 상관관계는 통계 연구에서—대중 매체와 과학 논문 모두에서 흔히 인용되는 지표입니다. 때때로 우리는 사전에 상관관계가 존재한다는 것을 알고 있으며, 그 상관관계의 강도를 찾고자 합니다. 우리는 86페이지의 “Reading Data from a CSV File”에서 파일에서 읽은 데이터 간의 상관관계를 계산할 때 이 예시를 확인할 수 있습니다. 

다른 경우에는 상관관계가 있을 수도 있다고 추측할 뿐이며, 실제로 상관관계가 있는지 확인하기 위해 데이터를 조사해야 합니다(다음 예시와 같이).

##### 텍스트 파일에서 데이터 읽기

Mydata.txt 파일을 사용하여, 이 장의 시작에서 고려한 기간 A 기간 동안의 기부 목록(한 줄당 하나씩)을 포함하겠습니다.

100 60 70 900 100 200 500 500 503 600 1000 1200

다음 프로그램이 이 파일을 읽고 파일에 저장된 숫자들의 합을 출력합니다:



```python
# 파일에 저장된 숫자들의 합을 찾으십시오 
def sum_data(filename): 
    s = 0
    with open(filename) as f:  #➊
        for line in f:         #➋
            s = s + float(line) #➌
    print('숫자의 합: {0}'.format(s))

if __name__ == '__main__':
    sum_data('mydata.txt')
```

Sum_data() 함수는 ➊에 있는 인수 filename으로 지정된 파일을 열고, 이를 한 줄씩 읽습니다(f는 파일 객체라고 하며, 열려 있는 파일을 가리키는 것으로 생각할 수 있습니다). ➋에서 우리는 float() 함수를 사용하여 각 숫자를 부동소수점 숫자로 변환하고, 모든 숫자를 읽을 때까지 계속 덧셈합니다. 마지막 숫자는 s라고 표시되어 있으며, 함수의 끝에 출력되는 숫자들의 합을 담고 있습니다.

프로그램을 실행하기 전에 먼저 적절한 데이터를 포함한 mydata.txt 파일을 생성하고 이를 프로그램과 동일한 디렉터리에 저장해야 합니다. File▸New Window를 클릭하고 새 창에 숫자를 한 줄에 하나씩 입력한 다음, 프로그램과 동일한 디렉터리에 mydata.txt 파일로 저장하면 IDLE에서 이 파일을 생성할 수 있습니다. 이제 프로그램을 실행하면 다음과 같은 출력이 나타납니다:

숫자의 합: 5733.0

이 장의 모든 프로그램은 입력 데이터가 목록에 포함되어 있다고 가정합니다. 파일의 데이터에 대한 이전 프로그램을 사용하려면 먼저 해당 데이터로 목록을 생성해야 합니다. 목록을 갖게 되면, 앞서 작성한 함수를 사용하여 해당 통계량을 계산할 수 있습니다. 다음 프로그램은 mydata.txt 파일에 저장된 숫자들의 평균을 계산합니다:

```python
''' 파일에 저장된 숫자들의 평균을 계산하기 '''
def read_data(filename): 
    numbers = [] 
    with open(filename) as f: 
        for line in f:  #➊
            numbers.append(float(line)) 
    return numbers 

def calculate_mean(numbers): 
    s = sum(numbers) 
    N = len(numbers) 
    mean = s/N 
    return mean 

if __name__ == '__main__':  #➋
    data = read_data('mydata.txt') 
    mean = calculate_mean(data) 
    print('평균: {0}'.format(mean))
```

Calculate_mean() 함수를 호출하기 전에 파일에 저장된 숫자를 읽어 목록으로 변환해야 합니다. 이를 수행하려면 파일을 한 줄로 읽는 read_data() 함수를 사용하십시오. 숫자를 합산하는 대신, 이 함수는 숫자를 부동소수점 숫자로 변환하여 리스트 번호 ➊에 추가합니다. 목록이 반환되며, 우리는 라벨 데이터 ➋로 이를 참조합니다. 그런 다음 우리는 데이터의 평균을 반환하는 calculate_mean() 함수를 호출합니다. 마침내 우리는 그것을 인쇄합니다.

프로그램을 실행하면 다음과 같은 출력이 나타납니다:

평균: 477.75

물론, 파일의 숫자가 이 예시와 다르면 평균값이 다르게 표시될 것입니다.

##### CSV 파일에서 데이터 읽기

쉼표로 구분된 값(CSV) 파일은 행과 열로 구성되며, 열은 서로 쉼표로 구분됩니다. 
운영 체제 또는 Microsoft Excel, OpenOffice Calc, LibreOffice Calc와 같은 특수 소프트웨어의 텍스트 편집기에서 CSV 파일을 사용할 수 있습니다.


다음은 몇 개의 숫자와 그 사각형이 포함된 샘플 CSV 파일입니다:
숫자,제곱
10,100
9,81
22,484

첫 번째 줄은 헤더라고 합니다. 이 경우, 파일의 첫 번째 열에 있는 항목은 숫자이며 두 번째 열에 있는 항목은 해당 사각형임을 알려줍니다. 다음 세 줄(또는 행)에는 숫자와 그 사각형이 쉼표로 구분되어 있습니다. 이 파일의 데이터를 .txt 파일에 대해 보여드린 것과 유사한 접근 방식을 사용하여 읽을 수 있습니다. 하지만 파이썬의 표준 라이브러리에는 CSV 파일을 읽고(쓰는) 전용 모듈(csv)이 있어 작업을 조금 더 쉽게 만들어 줍니다.

숫자와 그 제곱을 파일 numbers.csv에 저장하고, 프로그램과 같은 디렉터리에 보관하십시오. 다음 프로그램은 이 파일을 읽는 방법을 보여주며, 숫자를 제곱에 대해 표시하는 산점도를 생성합니다:

```python
import csv 
import matplotlib.pyplot as plt 

def scatter_plot(x, y): 
    plt.scatter(x, y) 
    plt.xlabel('Number') 
    plt.ylabel('Square') 
    plt.show() 

def read_csv(filename): 
    numbers = [] 
    squared = [] 
    with open(filename) as f: 
        reader = csv.reader(f) 
        next(reader)  # Skip the header row
        for row in reader: 
            numbers.append(int(row[0])) 
            squared.append(int(row[1])) 
    return numbers, squared 

if __name__ == '__main__': 
    numbers, squared = read_csv('numbers.csv') 
    scatter_plot(numbers, squared)
```

Read_csv() 함수는 프로그램 시작 시에 import되는 csv 모듈에 정의된 reader() 함수를 사용하여 CSV 파일을 읽습니다. 이 함수는 파일 객체 f가 인수 ➊로 전달된 경우 호출됩니다. 이 함수는 CSV 파일의 첫 번째 줄에 대한 포인터를 반환합니다. 우리는 파일의 첫 번째 줄이 헤더라는 것을 알고 있으며, 이를 건너뛰고자 하므로 next() 함수를 사용하여 포인터를 다음 줄로 이동합니다. 그런 다음 우리는 레이블 행 ➋에 의해 참조되는 파일의 모든 행을 읽고, 행[0]은 데이터의 첫 번째 열을 가리키고 행[1]은 두 번째를 의미합니다. 이 특정 파일에 대해서는 두 숫자가 모두 정수임을 알고 있으므로, 문자열에서 정수로 변환하기 위해 int() 함수를 사용하고 두 개의 리스트에 저장합니다. 그런 다음 목록이 반환됩니다—하나는 숫자를 포함하고 다른 하나는 사각형을 포함합니다.

그런 다음 두 개의 리스트를 사용하여 scatter_plot() 함수를 호출하여 산점도를 생성합니다. 앞서 작성한 find_corr_x_y() 함수는 두 집합의 숫자 사이의 상관계수를 쉽게 찾는 데에도 사용할 수 있습니다.

#1: 향상된 상관계수 찾기 프로그램

우리가 앞서 작성한 find_corr_x_y() 함수는 두 집합 사이의 상관 관계를 찾기 위해 두 집합의 길이가 동일하다고 가정합니다. 함수를 개선하여 먼저 목록의 길이를 확인하도록 하십시오. 두 값이 동일할 경우에만 함수는 남은 계산을 진행해야 합니다; 그렇지 않으면 상관관계를 찾을 수 없다는 오류 메시지를 출력해야 합니다.

```python
def find_corr_x_y(x,y): 
    n = len(x)  #label n 을 생성
    
    # products의 합을 찾습니다 
    prod = []
    for xi, yi in zip(x, y):    #➊
        prod.append(xi*yi)

    print ('prod: ', prod) 

    sum_prod_x_y = sum(prod)    #➋
    sum_x = sum(x)              #➌
    sum_y = sum(y)              #➍
    squared_sum_x = sum_x**2
    squared_sum_y = sum_y**2

    print ('squared_sum_x: ',squared_sum_x)
    print ('squared_sum_y: ',squared_sum_y)

    x_square = []
    for xi in x:        #➎
        x_square.append(xi**2)
    # 합을 찾습니다 
    x_square_sum = sum(x_square)
  
    y_square=[]
    for yi in y:
        y_square.append(yi**2)
    # 합을 찾습니다 
    y_square_sum = sum(y_square)
    # 상관계수를 계산 위한 공식 사용 
    numerator = n*sum_prod_x_y - sum_x*sum_y    #➏
    denominator_term1 = n*x_square_sum - squared_sum_x
    denominator_term2 = n*y_square_sum - squared_sum_y
    denominator = (denominator_term1*denominator_term2)**0.5    #➐
    correlation = numerator/denominator             #➑
    print ('상관계수: {0}'.format(correlation))
    return correlation
```

표는 등급을 두 등급으로 구분합니다: 1–6(1은 포함하지만 6은 제외)와 6–11(6은 포함하지만 11은 제외). 각 카테고리에 속하는 등급 수를 서로 비교하여 표시합니다. 각 클래스의 수와 각 클래스의 숫자 범위를 결정하는 것은 이 테이블을 만드는 데 필요한 두 가지 핵심 단계입니다. 이 예시에서는 각 클래스의 숫자 범위를 두 클래스에 균등하게 나누어 표시했습니다.

클래스 수를 임의로 선택할 수 있다고 가정하는 간단한 클래스 생성 방법은 다음과 같습니다:

```python
def create_classes(numbers, n): 
    low = min(numbers) 
    high = max(numbers) 
    # 각 클래스의 너비 
    width = (high - low)/n 
    classes = [] 
    a = low 
    b = low + width 
    classes = [] 
    while a < (high-width): 
        classes.append((a, b)) 
        a = b 
        b = a + width 
    # 마지막 클래스는 width보다 작을 수 있습니다. 
    classes.append((a, high+1)) 
    return classes
```

Create_classes() 함수는 두 개의 인수를 받습니다: 숫자 목록, 숫자, 그리고 생성할 클래스 수 n. 각 튜플이 클래스를 나타내는 튜플 목록을 반환합니다. 예를 들어, 숫자 7, 8, 9, 2, 10, 9, 9, 9, 9, 4, 5, 6, 1, 5, 6, 7, 8, 6, 1, 10으로 호출하고 n = 4인 경우, 다음 목록을 반환합니다: [(1, 3.25), (3.25, 5.5), (5.5), (5.5, 7.75), (7.75, 11)]. 목록을 확인하시면, 다음 단계는 각 숫자를 검토하여 반환된 클래스 중 어느 클래스에 속하는지 확인하는 것입니다.

##### 식의 인수분해 및 확장

Factor() 함수는 식을 인수로 분해하고, expand() 함수는 식을 확장하여 개별 항들의 합으로 표현합니다. 기본 대수 항등식 x2 – y2 = (x + y)(x – y)를 사용하여 이러한 함수들을 테스트해 보겠습니다. 왼쪽 정체성 [항등식의 왼쪽] 은 확장된 버전이며, 오른쪽은 해당 인수분해를 나타냅니다. 항등식에 두 개의 심볼이 있기 때문에 두 개의 Symbol 객체를 생성하겠습니다:

```python
from sympy import Symbol
x = Symbol('x')
y = Symbol('y')
```

다음으로, factor() 함수를 가져와 이를 사용하여 확장된 버전(항등식의 왼쪽에 있는)을 오른쪽에 있는 팩터화된 버전으로 변환합니다:

```python
from sympy import factor
expr = x**2 - y**2
factor(expr)
# (x - y)*(x + y)
```

예상대로, 우리는 식의 인수분해 버전을 얻게 됩니다. 이제 원본 확장 버전을 복구하기 위해 요인을 확장해 보겠습니다:

```python
from sympy import expand
factors = factor(expr)
expand(factors)
#x**2-y**2
```

우리는 인수분해된 식을 새로운 레이블인 factors에 저장한 다음, 이를 사용하여 expand() 함수를 호출합니다. 우리가 이것을 할 때, 우리는 처음에 사용한 원래 표현을 받게 됩니다. 더 복잡한 항등식 x3 + 3x2y + 3xy2 + y3 = (x + y)3을 사용해 보겠습니다:

```python
expr=x**3+3*x**2*y+3*x*y**2+y**3
factors=factor(expr)
factors
#(x+y)**3
expand(factors)
#x**3+3*x**2*y+3*x*y**2+y**3
```

Factor() 함수는 식을 인수분해할 수 있으며, 그 다음 expand() 함수는 인수분해된 식을 확장하여 원래 식으로 돌아갑니다.

가능한 인수분해가 불가능한 식에 대해 인수분해를 시도하면, 원래 식은 factor() 함수에 의해 반환됩니다. 예를 들어, 다음을 참조하십시오:

```python
expr = x + y + x*y
factor(expr)

#x*y + x + y
```

마찬가지로, expand()에 더 이상 확장할 수 없는 식을 전달하면 동일한 식을 반환합니다.

시리즈 인쇄

다음 시리즈를 고려하십시오:

사용자에게 숫자 n을 입력하도록 요청하고, 그 숫자에 대해 이 시리즈를 출력하는 프로그램을 작성합시다. 시리즈에서 x는 기호이며 n은 프로그램 사용자가 입력한 정수입니다. 이 급수의 n번째 항은 다음과 같습니다

다음 프로그램을 사용하여 이 시리즈를 인쇄할 수 있습니다:

```python
'''
시리즈를 출력합니다 : x + x**2 + x**3 + ... + x**n ____ _____ _____ 2 3 n 
''' 
from sympy import Symbol, pprint, init_printing  

def print_series(n): 
    # 역순으로 인쇄 시스템을 초기화합니다 
    init_printing(order='rev-lex') 
    x = Symbol('x')
    series = x                          #➊
    for i in range(2, n+1):             #➋ 
        series = series + (x**i)/i      #➌ 
    pprint(series) 
    
if __name__ == '__main__': 
    n = input('시리즈에 원하는 용어 수를 입력하십시오: ')
    print_series(int(n))                #➍ 
```

Print_series() 함수는 정수 n을 매개변수로 받아, 이는 시리즈에서 출력될 용어 수를 의미합니다. 함수를 ➍에서 호출할 때 입력을 정수로 변환한다는 점에 유의하십시오. 그런 다음 init_printing() 함수를 호출하여 시리즈를 역방향 사전 순서로 인쇄하도록 설정합니다.

➊에서 레이블과 시리즈를 생성하고 초기 값을 x로 설정합니다. 그런 다음, ➋에서 2부터 n까지의 정수를 순회하는 for 루프를 정의합니다. 루프가 반복될 때마다, ➌에서 각 항을 시리즈에 추가합니다. 다음과 같이:

i = 2, series = x + x**2 / 2
i = 3, series = x + x**2/2 + x**3/3

##### 값에서 대체하기

SymPy를 사용하여 값을 대수식에 삽입하는 방법을 살펴보겠습니다. 이렇게 하면 변수의 특정 값에 대한 식의 값을 계산할 수 있습니다. 다음과 같이 정의될 수 있는 수학식 x2 + 2xy + y2를 고려하십시오:



```python
from sympy import Symbol
x = Symbol('x')
y = Symbol('y')
x*x + x*y + x*y + y*y + y*y
#x**2 + 2*x*y + y**2
```

파이썬 사전

사전은 파이썬에서 또 다른 유형의 데이터 구조이며(리스트와 튜플은 데이터 구조의 다른 예이며, 앞서 보셨듯이). 사전은 중괄호 안에 키‐값 쌍을 포함하고 있으며, 각 키는 콜론으로 구분된 값과 일치합니다. 앞선 코드 목록에서 우리는 사전 {x:1, y:2}를 subs() 메서드에 대한 인수로 입력했습니다. 이 사전은 x:1과 y:2라는 두 개의 키‐값 쌍을 가지고 있으며, 여기서 x와 y는 키이고 1과 2는 해당 값. 괄호 안에 연관된 키를 입력하면 사전에서 값을 가져올 수 있으며, 이는 인덱스를 사용하여 목록에서 요소를 가져오는 것과 같습니다. 예를 들어, 여기서는 간단한 사전을 만든 다음 key1에 해당하는 값을 가져옵니다:


```python
sampledict = {"key1": 5, "key2": 20}
sampledict
#["key1"]5, "key2": 20}
#사전에 대해 자세히 알아보려면 부록 B를 참조하십시오.
```

결과를 더 간소화하고 싶다면—예를 들어 서로 상쇄되는 용어가 있는 경우, SymPy의 simplify() 함수를 다음과 같이 사용할 수 있습니다.

➊ >>> expr_subs = expr.subs({x:1-y}) 
>>> from sympy import simplify
➋ >>> simplify(expr_subs) 1

➊에서 우리는 식에서 x = 1 – y를 대입한 결과를 나타내기 위해 새로운 레이블 expr_subs를 생성합니다. 그런 다음 SymPy에서 simplify() 함수를 가져와서 ➋에서 호출합니다. 결과는 1이 되는 이유는 식의 다른 항들이 서로 상쇄하기 때문입니다.

앞 예제에 표현식의 간소화된 버전이 있었지만, simplify() 함수를 사용하여 SymPy에 이를 간소화하도록 요청해야 했습니다. 다시 말하지만, 이는 SymPy가 요청받지 않으면 어떠한 단순화도 하지 않기 때문입니다.

Simplify() 함수는 로그와 삼각함수를 포함하는 복잡한 식을 단순화할 수도 있지만, 여기서는 이에 대해 설명하지 않겠습니다.

##### 시리즈의 가치 계산

시리즈 인쇄 프로그램을 다시 살펴보겠습니다. 시리즈를 출력하는 것 외에도, 우리 프로그램이 x의 특정 값에 대한 시리즈의 값을 찾을 수 있기를 원합니다. 즉, 이제 우리 프로그램은 사용자로부터 두 가지 입력을 받게 됩니다—시리즈 내 항의 수와 시리즈 값이 계산되는 x의 값입니다. 그럼 프로그램은 수열과 합을 모두 출력합니다. 다음 프로그램은 시리즈 인쇄 프로그램을 확장하여 이러한 개선 사항을 포함합니다:

```python
'''
시리즈를 출력합니다: 
x + x**2 + x**3 + ... + x**n 
____ _____ _____ 
2 3 n 
''' 
from sympy import Symbol, pprint, init_printing 
def print_series(n, x_value): 
    # 역순으로 인쇄 시스템을 초기화합니다 
    init_printing(order='rev-lex') 
    x = Symbol('x') 
    series = x 
    for i in range(2, n+1): 
        series = series + (x**i)/i 
    pprint(series)
    # x_value에서 시리즈를 평가하십시오
    series_value = series.subs({x:x_value}) #➊ 
    print('시리즈의 {0}에서의 값: {1}'.format(x_value, series_value)) 

if __name__ == '__main__': 
    n = input('시리즈에 포함하려는 용어 수를 입력하십시오: ')
    x_value = input('시리즈를 평가하고자 하는 x의 값을 입력하십시오: ') #➋ 
    print_series(int(n), float(x_value))
```

Print_series() 함수는 이제 추가 인수인 x_value를 받아, 시리즈를 평가해야 하는 x의 값을 의미합니다. ➊에서 우리는 subs() 메서드를 사용하여 평가를 수행하고, 레이블 series_value를 사용하여 결과를 참조합니다. 다음 줄에 결과를 표시합니다.

➋에 있는 추가 입력문은 사용자가 x_value 레이블을 사용하여 x 값을 입력하도록 요청합니다. Print_series() 함수를 호출하기 전에, float() 함수를 사용하여 이 값을 부동소수점 등가값으로 변환합니다.

프로그램을 지금 실행하면 두 입력을 입력하고 시리즈와 시리즈 값을 출력합니다:

```python
''' 두 표현식의 곱 ''' 
from sympy import expand, sympify 
from sympy.core.sympify import SympifyError 

def product(expr1, expr2): 
    prod = expand(expr1*expr2) 
    print(prod) 
if __name__=='__main__':
    expr1 = input('첫 번째 식을 입력하세요: ') #➊ 
    expr2 = input('두 번째 표현식 입력: ') #➋ 
        
    try: 
        expr1 = sympify(expr1) 
        expr2 = sympify(expr2) 
    except SympifyError: 
        print('Invalid input') 
    else: 
        product(expr1, expr2) #➌
```

방정식 풀기

SymPy의 solve() 함수는 방정식의 해를 찾는 데 사용할 수 있습니다. X와 같이 변수를 나타내는 기호가 포함된 식을 입력하면, solve()가 해당 기호의 값을 계산합니다. 이 함수는 입력한 식이 0이라고 가정하여 항상 계산을 수행합니다—즉, 기호를 대입했을 때 전체 식이 0이 되는 값을 출력합니다. 간단한 방정식 x – 5 = 7부터 시작하겠습니다. Solve()를 사용하여 x의 값을 찾고자 한다면, 먼저 방정식의 한쪽을 0으로 만들어야 합니다 (x – 5 – 7 = 0). 그럼, 우리는 다음과 같이 solve()를 사용할 준비가 되었습니다:



```python
from sympy import Symbol, solve
x = Symbol('x')
expr = x - 5 - 7
solve(expr)
#[12]

```

Solve()를 사용할 때, 'x'의 값을 12로 계산합니다. 이는 식 (x – 5 – 7)을 0으로 만들기 때문입니다.

결과 12가 목록에 반환된다는 점에 유의하십시오. 방정식은 여러 개의 해를 가질 수 있습니다—예를 들어, 이차 방정식은 두 개의 해를 가질 수 있습니다. 그 경우, 목록은 모든 솔루션을 구성원으로 갖게 됩니다. 또한 solve() 함수에 결과를 반환하도록 요청하여 각 멤버가 대신 사전이 될 수 있습니다. 각 사전은 기호(변수 이름)와 그 값(해답)으로 구성됩니다. 이는 두 개 이상의 변수를 풀어야 하는 동시 방정식을 풀 때 특히 유용합니다. 왜냐하면 해가 사전 형태로 반환될 때 어떤 해가 어느 변수에 해당하는지 알 수 있기 때문입니다.

##### 이차 방정식 풀기

1장에서는 두 근에 대한 공식을 작성하고 상수 a, b, c의 값을 대입함으로써 이차 방정식 ax2 + b + c = 0의 근을 찾았습니다. 이제 우리는 SymPy의 solve() 함수를 사용하여 수식을 작성하지 않고도 근을 찾는 방법을 배워보겠습니다. 예시를 살펴보겠습니다:

```python
from sympy import solve #➊ >>> 
x = Symbol('x')
expr = x**2 + 5*x + 4   #➋ >>> 
solve(expr, dict=True)  #➌ >>> 
# [{x: -4}, {x: -1}]      ➍ 
```

Solve() 함수는 먼저 ➊에서 가져옵니다. 그런 다음 우리는 기호 x와 이차 방정식 x**2 + 5*x + 4에 해당하는 식을 정의합니다, ➋. 그런 다음, 앞의 식이 ➌에 있는 solve() 함수를 호출합니다. Solve() 함수의 두 번째 인수(dict=True)는 결과를 Python 사전 목록으로 반환하고자 함을 지정합니다.

반환된 목록의 각 해는 해당 기호를 해당 값과 일치하는 키로 사용하는 사전입니다. 해답이 비어 있으면 빈 목록이 반환됩니다. 앞선 방정식의 근은 ➍에서 볼 수 있는 바와 같이 –4와 –1입니다.

우리는 첫 번째 장에서 방정식의 근을 알게 되었습니다.

X2 + x + 1 = 0

복소수입니다. Solve()를 사용하여 찾아봅시다:

```python
x=Symbol('x')
expr = x**2 + x + 1
solve(expr, dict=True)
#[{x: -1/2 - sqrt(3)*I/2}, {x: -1/2 + sqrt(3)*I/2}]
```

두 근은 모두 허구이며, I 기호로 표시된 허구 성분으로 예상됩니다.

##### 다른 변수와 관련된 하나의 변수 풀이

방정식의 근을 찾는 것 외에도, 우리는 심볼릭 수학을 활용하여 solve() 함수를 이용해 방정식의 한 변수를 다른 변수들을 기준으로 표현할 수 있습니다. 일반적인 이차 방정식 ax2 + bx + c = 0의 근을 찾는 방법을 살펴보겠습니다. 이를 위해 우리는 x와 세 가지 추가 기호인 a, b, c를 정의할 것이며, 이는 세 개의 상수에 해당합니다:



```python
x = Symbol('x')
a = Symbol('a')
b = Symbol('b')
c = Symbol('c')
```

다음으로, 방정식에 해당하는 식을 작성하고 그 위에 solve() 함수를 사용합니다:

```python
expr = a*x*x + b*x + c
solve(expr, x, dict=True)
# [{x: (-b + sqrt(-4*a*c + b**2))/(2*a)}, {x: -(b + sqrt(-4*a*c + b**2))/(2*a)}]
```

여기서는 solve() 함수에 추가 인수 x를 포함해야 합니다. 방정식에 심볼이 하나 이상 존재하므로, solve()에 어떤 심볼을 풀어야 하는지 알려야 하며, 이는 두 번째 인수로 x를 전달함으로써 이를 나타냅니다. 예상대로, solve()는 이차 공식을 출력합니다: 값을 찾기 위한 일반 공식입니다.

명확히 말씀드리면, 두 개 이상의 심볼을 가진 방정식에 solve()를 사용할 때는 두 번째 인수로 해결할 심볼을 지정합니다(그리고 이제 세 번째 인수는 결과가 반환되는 방식을 지정합니다).

```python
from sympy import Symbol, solve, pprint
s = Symbol('s')
u = Symbol('u')
t = Symbol('t')
a = Symbol('a')
expr = u*t + (1/2)*a*t*t - s
t_expr = solve(expr,t, dict=True)
pprint(t_expr)

# 결과는 다음과 같습니다:
```
